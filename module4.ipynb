```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
data = pd.read_csv('bank_marketing_test.csv')

# Displaying the statistical summary of the dataset
summary = data.describe()
print(summary)
```

```python
# 2. Data Elaboration
# Select only numeric columns for analysis
numeric_data = data.select_dtypes(include=[np.number])

# Calculate additional statistical measures
data_summary = pd.DataFrame({
    'Mean': numeric_data.mean(),
    'Median': numeric_data.median(),
    'Variance': numeric_data.var(),
    'Standard Deviation': numeric_data.std(),
    'Skewness': numeric_data.skew(),
    'Kurtosis': numeric_data.kurt()
})
print(data_summary)
```

```python
# 3. 1-D Statistical Data Analysis
# Distribution plot for 'age'
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=20, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 4. 2-D Statistical Data Analysis
# Scatter plot for 'duration' vs 'age'
plt.figure(figsize=(8, 5))
plt.scatter(data['duration'], data['age'], color='red')
plt.title('Scatter Plot: Duration vs Age')
plt.xlabel('Duration of Last Contact (seconds)')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
# 5. n-D Statistical Data Analysis
# Pair plot for multiple variables
sns.pairplot(data[['age', 'duration', 'campaign', 'pdays']], diag_kind='kde')
plt.suptitle('Pair Plot of Multiple Variables', y=1.02)
plt.show()
```

```python
# Selecting only numeric columns for statistical analysis
numeric_data = data.select_dtypes(include=[np.number])

# Displaying the first few rows of numeric data
print(numeric_data.head())
```

```python
# Calculating additional statistical measures for numeric columns
numeric_data_summary = pd.DataFrame({
    'Mean': numeric_data.mean(),
    'Median': numeric_data.median(),
    'Variance': numeric_data.var(),
    'Standard Deviation': numeric_data.std(),
    'Skewness': numeric_data.skew(),
    'Kurtosis': numeric_data.kurt()
})

# Displaying the detailed summary for numeric columns
print(numeric_data_summary)
```

```python
# Encoding categorical columns
encoded_data = data.copy()

# Encoding 'job'
encoded_data['job'] = encoded_data['job'].astype('category').cat.codes

# Encoding 'marital'
encoded_data['marital'] = encoded_data['marital'].map({'married': 0, 'single': 1, 'divorced': 2})

# Encoding 'education'
education_mapping = {
    'unknown': 0, 'basic.4y': 1, 'basic.6y': 2, 'basic.9y': 3,
    'high.school': 4, 'professional.course': 5, 'university.degree': 6
}
encoded_data['education'] = encoded_data['education'].map(education_mapping)

# Encoding 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome', 'y'
categorical_mappings = {
    'default': {'no': 0, 'yes': 1, 'unknown': 2},
    'housing': {'no': 0, 'yes': 1},
    'loan': {'no': 0, 'yes': 1},
    'contact': {'telephone': 0, 'cellular': 1, 'unknown': 2},
    'month': {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
              'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12},
    'day_of_week': {'mon': 1, 'tue': 2, 'wed': 3, 'thu': 4, 'fri': 5},
    'poutcome': {'nonexistent': 0, 'failure': 1, 'success': 2},
    'y': {'no': 0, 'yes': 1}
}

for col, mapping in categorical_mappings.items():
    encoded_data[col] = encoded_data[col].map(mapping)

# Check if all columns are numeric now
print(encoded_data.dtypes)
```

```python
# Now calculating statistical measures for encoded data
encoded_data_summary = pd.DataFrame({
    'Mean': encoded_data.mean(),
    'Median': encoded_data.median(),
    'Variance': encoded_data.var(),
    'Standard Deviation': encoded_data.std(),
    'Skewness': encoded_data.skew(),
    'Kurtosis': encoded_data.kurt()
})

# Display the summary
print(encoded_data_summary)
```

```python
# Pair Plot with Encoded Data
sns.pairplot(encoded_data[['age', 'duration', 'campaign', 'education', 'housing', 'y']], hue='y', palette='pastel')
plt.suptitle('Pair Plot with Encoded Categorical Variables', y=1.02)
plt.show()
```

```python
# Selecting relevant columns for correlation analysis
correlation_columns = ['age', 'duration', 'campaign', 'education', 'housing', 'y']

# Creating the heatmap
correlation_matrix = encoded_data[correlation_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Selected Variables')
plt.show()
```
